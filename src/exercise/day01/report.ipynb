```python
# system prompt에 출력 형식을 지정

import os
import openai
from pykrx import stock
import pandas as pd


# 삼성전자의 주식 코드
stock_code = "005930"

# 특정 기간 동안의 주가 데이터를 가져오기
start_date = "20230301"  # 시작 날짜 예시
end_date = "20230331"    # 종료 날짜 예시
df = stock.get_market_ohlcv_by_date(start_date, end_date, stock_code)


openai.api_key = os.getenv("OPENAI_API_KEY")

response = openai.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {
      "role": "system",
      "content": "당신은 주식 종목 분석가입니다. system에 주어진 종목 정보 데이터만 이용해 사용자에게 정보를 제공해야 합니다. 답변에 항상 '삼성전자' 라고 종목 이름을 이야기해야 합니다."
    },
    {
      "role": "system",
      "content": df.to_json()
    },
    {
        "role": "user",
        "content": "앞으로 주가가 오를꺼 같아?"
    }
  ],
  temperature=0,
  max_tokens=256,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)

print(response.choices[0].message.content)
```

```python
# system prompt에 출력 형식을 지정
import openai
import requests

# API URL 예시
url = 'http://apis.data.go.kr/B551011/GoCamping/basedList'

# API에 보낼 파라미터 예시
params = {
    'MobileOS': 'ETC',
    'MobileApp': 'app',
    'serviceKey': 'Pw33fYa370vIl1aY4QA5dDh4FXFGlJWhNblfgwYD2Ltbnk8aaZGNzPIlPoZbSR0S7mzuZnIHGbvpyI2acL7uKQ==',
    '_type': 'json',
}

# GET 요청을 보내고 응답을 받음
data = requests.get(url, params=params, verify=False)


openai.api_key = os.getenv("OPENAI_API_KEY")

response = openai.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {
      "role": "system",
      "content": "당신은 캠핑 추천 전문가입니다. system에 주어진 데이터만 이용해 사용자에게 정보를 제공해야 합니다."
    },
    {
      "role": "system",
      "content": data.text
    },
    {
        "role": "user",
        "content": "강원도 여름에 가지 좋은 캠핑장 추천해줘"
    }
  ],
  temperature=0,
  max_tokens=256,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)

print(response.choices[0].message.content)
```
